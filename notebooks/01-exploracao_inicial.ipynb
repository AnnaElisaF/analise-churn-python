```python
import pandas as pd
import numpy as np

```

```python
# Carregando o dataset a partir do arquivo CSV
# Usamos '../' para "subir um nível" da pasta 'notebooks' para a raiz do projeto e depois entrar na pasta 'data'
caminho_arquivo = '../data/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(caminho_arquivo)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df.info()
```

```python
df[df['TotalCharges'].isnull()]
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)

df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

ax = sns.countplot(x='Churn', data=df, palette='viridis')

ax.set_title('Distribuição de Clientes por Status de Churn', fontsize=16)

ax.set_xlabel('Cliente Cancelou o Serviço? (Churn)', fontsize=12)
ax.set_ylabel('Quantidade de Clientes', fontsize=12)

plt.show()
```

```python
# Calculando a contagem exata de cada categoria na coluna 'Churn'
churn_counts = df['Churn'].value_counts()

# Usando normalize=True para obter a proporção em vez da contagem
churn_percentage = df['Churn'].value_counts(normalize=True) * 100

print("--- Análise da Variável 'Churn' ---")
print("\nContagem Absoluta:")
print(churn_counts)
print("\nPorcentagem (%):")
print(f'{churn_percentage}')
```

```python
plt.figure(figsize=(10, 6))

ax = sns.countplot(x='Contract', hue='Churn', data=df, palette='viridis')

ax.set_title('Relação entre tipo de contrato e churn', fontsize=16)
ax.set_xlabel('Tipo de contrato', fontsize=12)
ax.set_ylabel('Quantidade de Clientes', fontsize=12)

plt.show()
```

```python
colunas_categoricas = df.select_dtypes(include=['object']).drop('customerID', axis=1)

for coluna in colunas_categoricas:
    plt.figure(figsize=(10, 7))
    ax = sns.countplot(x=coluna, hue='Churn', data=df, palette='viridis')

    ax.set_title(f'Relação entre {coluna} e churn', fontsize=16)
    ax.set_xlabel(coluna, fontsize=12)
    ax.set_ylabel('Quantidade de clientes', fontsize=12)

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
```

```python
colunas_numericas = ['tenure', 'MonthlyCharges', 'TotalCharges']

for coluna in colunas_numericas:
    plt.figure(figsize=(10, 7))
    sns.histplot(x=coluna, hue='Churn', data=df, multiple='stack', kde=True, palette='viridis')

    plt.title(f'Distribuição de {coluna} por Churn', fontsize=16)
    plt.xlabel(coluna, fontsize=12)
    plt.ylabel('Densidade de clientes', fontsize=12)
    plt.show()

for coluna in colunas_numericas:
    plt.figure(figsize=(10, 7))

    sns.boxplot(x='Churn', y=coluna, data=df, palette='viridis')

    plt.title(f'Boxplot de {coluna} por Churn', fontsize=16)
    plt.xlabel('Cliente cancelou o serviço? (Churn)', fontsize=12)
    plt.ylabel(coluna, fontsize=12)
    plt.show()
```

```python
df_processado = df.copy()

df_processado = df_processado.drop('customerID', axis=1)
df_processado.head()
```

```python
df_processado = pd.get_dummies(df_processado, drop_first=True)

print('Dataframe após one-hot encoding')
display(df_processado.head())

print('Informações novo dataframe')
df_processado.info()
```

```python
X = df_processado.drop('Churn_Yes', axis=1)
y = df_processado['Churn_Yes']

print('Dimensões de x (features):', X.shape)
print('Dimensões de y (alvo)', y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("\nDimensões dos conjuntos de dados:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print('\nDados escalonados com sucesso!')
```

```python
# Modelo 1: Regressão Logística

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

log_reg = LogisticRegression(random_state=42)

print('Treinando o modelo de Regressão Logística...')
log_reg.fit(X_train_scaled, y_train)
print('Modelo treinado com sucesso.')

y_pred_log_reg = log_reg.predict(X_test_scaled)

print('\nResultados da avaliação do modelo de regresão logística')

accuracy = accuracy_score(y_test, y_pred_log_reg)
print(f'Acurácia: {accuracy * 100:.2f}%')

print('matriz de confusão')
cm = confusion_matrix(y_test, y_pred_log_reg)
print(cm)

print('\nRelatório de Classificação:')
print(classification_report(y_test, y_pred_log_reg, target_names=['Não Churn (0)', 'Churn (1)']))
```

### Tabela de Comparação de Modelos

| Modelo                | Acurácia | F1-Score (Churn) | Recall (Churn) |
| :-------------------- | :------- | :--------------- | :------------- |
| Regressão Logística   | 0.79     | 0.56             | 0.52           |


```python
# Modelo 2: Árvore de decisão

from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42)

print('Treinando o modelo...')
tree_clf.fit(X_train_scaled, y_train)
print('Modelo treinado com sucesso!')

y_pred_tree = tree_clf.predict(X_test_scaled)

print('\nResultados da avaliação')
accuracy_tree = accuracy_score(y_test, y_pred_tree)
print(f'Acurácia: {accuracy_tree * 100:.2f}%')

print('Matriz de confusão')
print(confusion_matrix(y_test, y_pred_tree))

print('\nRelatório de classificação: ')
print(classification_report(y_test, y_pred_tree, target_names=['Não Churn (0)', 'Churn (1)']))
```

```python
# Modelo 3: Random Forest

from sklearn.tree import plot_tree

plt.figure(figsize=(25, 12))

plot_tree(
    tree_clf,
    max_depth=3,
    feature_names=X.columns,
    class_names=['Não Churn', 'Churn'],
    filled=True,
    rounded=True,
    fontsize=10
)

plt.title('Visualização da Árvore de Decisão (3 primeiros níveis)', fontsize=20)
plt.show()
```

### Tabela de Comparação de Modelos

| Modelo                | Acurácia | F1-Score (Churn) | Recall (Churn) |
| :-------------------- | :------- | :--------------- | :------------- |
|   Árvore de Decisão   | 0.73     | 0.51             | 0.53           |

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=1)

print('Treinando modelo')
rf_clf.fit(X_train_scaled, y_train)
print('Modelo treinado com sucesso!')

y_pred_rf = rf_clf.predict(X_test_scaled)

print('\nResultados da avaliação')
# Acurácia
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Acurácia: {accuracy_rf * 100:.2f}%")

# Matriz de Confusão
print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, y_pred_rf))

# Relatório de Classificação
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_rf, target_names=['Não Churn (0)', 'Churn (1)']))
```

### Tabela de Comparação de Modelos

| Modelo                | Acurácia | F1-Score (Churn) | Recall (Churn) |
| :-------------------- | :------- | :--------------- | :------------- |
|   Random Forest       | 0.78     | 0.54             | 0.47           |

```python
feature_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_clf.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(
    x='importance',
    y='feature',
    data=feature_importance_df.head(15),
    palette='viridis'
)

plt.title('Top 15 características mais importantes - Random Forest', fontsize=16)
plt.xlabel('Importância', fontsize=12)
plt.ylabel('Características', fontsize=12)
plt.show()
```
